```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import copy
def linear_reg(x,w,b):
  f=np.dot(x,w)+b
  return f

def cost_func(x,y,w,b,lambda_=1):
  m=x.shape[0]
  cost=0
  for i in range(m):
    f=np.dot(w,x[i])+b
    cost=cost+(f-y[i])**2
  total_cost = 1 / (2 * m) * cost
  reg_cost = 0
  for j in range(len(w)):
    reg_cost += (w[j] ** 2)
  total_cost = total_cost + (lambda_ / (2 * m)) * reg_cost
  return total_cost
def gradeant(x,y,w,b,lambda_):
  m,n=x.shape
  df_dw=np.zeros((n,))
  df_db=0
  for i in range(m):
    err=(np.dot(x[i],w)+b)-y[i]
    for j in range(n):
      df_dw[j]=df_dw[j]+err * x[i,j]
    df_db=df_db+err

  df_dw=df_dw/m
  df_db=df_db/m
  for j in range(n):
    df_dw[j]=df_dw[j]+(lambda_ * w[j]) / m
  return df_dw,df_db

def gradient_descent(x, y, w_in, b_in, alpha, num_iters, gradient_function,lambda_):
  b=b_in
  w=copy.deepcopy(w_in)

  for i in range(num_iters):
    df_dw,df_db=gradient_function(x,y,w,b,lambda_)
    w=w-alpha*df_dw
    b=b-alpha*df_db
  return w,b

x = np.array([
    [1.0, 2],
    [2.0, 3],
    [3.0, 3],
    [4.0, 5],
    [5.0, 4]
])


y = np.array([100, 150, 200, 250, 300])
m,n=x.shape
w=np.zeros((n,))
b=0
alpha=0.01
num_iters=1000
lambda_=0.2
w_final,b_final=gradient_descent(x,y,w,b,alpha,num_iters,gradeant,lambda_)
print(f"w: {w_final}, b:{b_final}")
x_test = np.array([2, 3])
print(f"the price of house {linear_reg(x_test,w_final,b_final)}")
print(f"the cost of house {cost_func(x,y,w_final,b_final)}")
```
